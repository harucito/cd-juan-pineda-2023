```python
import itertools
import statsmodels.api as sm
import pandas as pd

# Supongamos que tienes un DataFrame llamado 'df' con las variables independientes en las columnas 2 a 10 y la variable dependiente 'Survived'.
x = df.iloc[:, [2, 3, 4, 5, 6, 7, 8, 9, 10]]
y = df['Survived']

# Crear una lista de todas las combinaciones posibles de variables, desde 1 hasta n variables
variables = list(x.columns)
combinations = []

for i in range(1, len(variables) + 1):
    combinations.extend(itertools.combinations(variables, i))

best_pseudo_r2 = -1  # Inicializar la variable para almacenar el mejor pseudo R^2
best_model = None    # Inicializar la variable para almacenar el mejor modelo
best_features = None  # Inicializar la variable para almacenar las mejores variables

# Iterar a través de las combinaciones y ajustar modelos de regresión logística
for combo in combinations:
    selected_features = list(combo)
    x_subset = x[selected_features]

    logit_model = sm.Logit(y, x_subset)
    result = logit_model.fit()

    pseudo_r2 = result.prsquared  # Obtener el pseudo R^2

    if pseudo_r2 > best_pseudo_r2:
        best_pseudo_r2 = pseudo_r2
        best_model = result
        best_features = selected_features

    # Imprimir las variables y el pseudo R^2 en esta iteración
    print("Variables:", selected_features)
    print("Pseudo R^2:", pseudo_r2)
    print("=" * 30)

# Imprimir el resumen del mejor modelo
print("Mejor modelo seleccionado con variables:", best_features)
print(best_model.summary())
print("Pseudo R^2 del mejor modelo:", best_pseudo_r2)
```
